```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [2**2, 2**3, 2**4, 2**5]
kd = [0.06, 0.23, 0.98, 3.35]
my = [0.06, 0.27, 0.97, 2.78]
```

```python
print("X values: ", x)
print("Average KD-TREE search times: ", kd)
print("Average MY-TREE search times:", my)
plt.plot(x, kd, 
         label="KD-TREE", 
         linewidth=2
         
        )
plt.scatter(x, kd,
           color='b',
            alpha=0.5
           )
plt.scatter(x, my,
           color='orange',
            alpha=0.5
           )
plt.plot(x, my, 
         label="MY-TREE", 
         linewidth=2
        )
plt.xlabel("QUERY RANGE")
plt.ylabel("TIME (SECONDS)")
plt.legend(fontsize=10)
plt.title("TREE SEARCH TIME PER QUERY RANGE")
plt.figure(figsize=(10,10))
plt.show();
```
